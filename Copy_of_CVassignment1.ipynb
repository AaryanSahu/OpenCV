## Question 1:
Plot the following distributions:
- Normal distribution with μ=0, σ<sup>2</sup>=1
- Normal distribution with μ=4, σ<sup>2</sup>=5
- Uniform distribution with a=1, b=5.

```
# Add Question 1 code here
import numpy as np
import matplotlib.pyplot as plt

def f(x,mean,variance):
 return np.exp(-np.square(x-mean)/2*variance)/(np.sqrt(2*np.pi*variance))

x = np.arange(-5,5,.01)
y=f(x,0,1)
z=f(x,4,5)


plt.plot(x,y,color='g',label='Graph1')
plt.plot(x,z,color='b',label='Graph2')
plt.legend()
plt.ylabel('Normal distribution')
plt.show()



```
